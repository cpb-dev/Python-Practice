# Group B Stats
**England, IR Iran, USA, Wales**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os

England = pd.read_csv('../input/world-cup-2022-national-teams-data-set/2022 England.csv')
Iran = pd.read_csv('../input/world-cup-2022-national-teams-data-set/2022 IR Iran.csv')
Usa = pd.read_csv('../input/world-cup-2022-national-teams-data-set/2022 United States.csv')
Wales = pd.read_csv('../input/world-cup-2022-national-teams-data-set/2022 Wales.csv')
```

```python
e_missing_vals = (England.isnull().sum())
i_missing_vals = (Iran.isnull().sum())
u_missing_vals = (Usa.isnull().sum())
w_missing_vals = (Wales.isnull().sum())

#Camparing what cells are missing values to establish better data understanding
print("England's Data \n", England.shape, e_missing_vals[e_missing_vals > 0], "\n\n")
print("IR Iran's Data \n", Iran.shape,  i_missing_vals[e_missing_vals > 0], "\n\n")
print("USA's Data \n", Usa.shape,  u_missing_vals[e_missing_vals > 0], "\n\n")
print("Wales' Data \n", Wales.shape,  w_missing_vals[e_missing_vals > 0], "\n\n")
```

Within most of the data sets there are missing values for some of the players.
This seems to be even and consistant for a particular player, as each column is missing the same number of values it's evident they are within a full players row. 

These rows need to be dropped depending on important cells of information, such as playing_time. This cell is important as it shows stats have actually been collected can make a player useful for predictions, without this data they could be a hinderance.

```python
#Dropping the values from the datasets where information loss is found
print("England data = ", England.shape, "IR Iran data = ", Iran.shape, "USA data = ", Usa.shape, "Wales data = ", Wales.shape)

England = England.dropna(subset=['Min_Playing_Time'])
Iran = Iran.dropna(subset=['Min_Playing_Time'])
Usa = Usa.dropna(subset=['Min_Playing_Time'])
Wales = Wales.dropna(subset=['Min_Playing_Time'])

print("England data = ", England.shape, "IR Iran data = ", Iran.shape, "USA data = ", Usa.shape, "Wales data = ", Wales.shape)
```

```python
cols = England.columns.tolist()
print(cols)
print(len(cols))

#Listing the important stats for visualisation
imp_cols = [7, 11, 12, 16, 17]
```

# England Stats

```python
for col in imp_cols:
    data = England.sort_values(cols[col])
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(data['Player'], data[cols[col]], color="violet")
    ax.set_xlabel(cols[col])
    plt.xticks(rotation = 90)
    ax.set_title(cols[col])
    plt.show()
    
```
